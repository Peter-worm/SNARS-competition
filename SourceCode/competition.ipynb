```python
import numpy as np
import random
from collections import defaultdict
import csv
from sklearn.metrics import adjusted_rand_score
from unk_greedy import find_partition
from utils import draw_reduced, map_from_1
import networkx as nx


import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict




def check_u(partition, blocksize=5):
    target = sorted(list(range(len(partition)//blocksize)) * blocksize)
    return adjusted_rand_score(np.array(target), np.array([partition[k] for k in range(len(partition))]))
    


with open("competition_data/D1-UNC.csv", 'r') as file:
    reader = csv.reader(file)
    adj_matrix = np.array([list(map(int, row)) for row in reader])

partition = find_partition(adj_matrix, 1.0)
g = nx.from_numpy_array(adj_matrix)

print(check_u(partition, 5))
result = map_from_1(partition)
with open("../results/D1-UNC.csv", "w+") as file:
    for key, value in result.items():
        file.write(f"{key},{value}\n")



draw_reduced(g, partition)
```
